```python
import math
```

```python
#Given a range of numbers between m and n
#Find the first pair of prime numbers -

def gap(g,m,n):
    lastPrime=-1
    for i in range(m,n+1):
        if all(i%x!=0 for x in range(2,int(math.sqrt(i))+1)):
            if i-g == lastPrime:
                return [i-g,i] 
            lastPrime=i
    return None

gap(2,100,110)
#gap(2,100,110)       
```
