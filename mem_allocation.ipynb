```python
def mem_alloc(banks):
    past=[]
    length=len(banks)
    c=0
    while banks not in past:
        past.append(banks)
        banks=banks.copy()
        
        c+=1
        re=max(banks)
        maxBank=banks.index(re)
        banks[maxBank]=0
        counter=1
        while re>0:
            banks[(maxBank+counter)%(length)]+=1
            re-=1
            counter+=1
    return past

mem=[5, 1, 10, 0, 1, 7, 13, 14, 3, 12, 8, 10, 7, 12, 0, 600]
#mem=[0,2,7,0]
#mem=[5, 1, 10, 0, 1, 7, 13, 14, 3, 12, 8, 10, 7, 12, 0, 6]
mem_alloc(mem)
```
